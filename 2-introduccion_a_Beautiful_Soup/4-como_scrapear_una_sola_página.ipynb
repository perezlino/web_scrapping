### **Cómo scrapear una sola página**

Nos tenemos que guiar según la **`estructura de árbol HTML`** para scrapear una página web:

1. ID
2. Class name
3. Tag name, CSS selector
4. XPath 

```python
from bs4 import BeautifulSoup
import requests

website = 'https://subslikescript.com/movie/Titanic-120338'
result = requests.get(website)
content = result.text

soup = BeautifulSoup(content, 'lxml')

print(soup.prettify())
```

```python
box = soup.find('article', class_="main-article")

title = box.find('h1').get_text()

title
```

```python
transcript = box.find('div', class_="full-script").get_text()

transcript
```

Con **`strip`** eliminamos los espacios en blanco de la izquierda como el de la derecha que tenga cada palabra. Con **`separator`** generamos un espacio entre cada palabra.

```python
transcript = box.find('div', class_="full-script").get_text(strip=True, separator=' ')

transcript
```
